## Parte 1. Consultas básicas de exploración
¿ Cuántos viajes hay en total en el dataset NYC Taxis Trips?

```python
%sql
SELECT count(*) FROM samples.nyctaxi.trips;
```

¿ Qué columnas tiene la tabla samples.nyctaxi.trips?

```python
%sql 
DESCRIBE samples.nyctaxi.trips;
```

¿Cuál es el viaje más largo en el dataset? y el más corto ?, exclutendo los ceros.


```python
%sql 
SELECT MAX(samples.nyctaxi.trips.trip_distance) AS viaje_mas_largo, MIN(samples.nyctaxi.trips.trip_distance) AS viaje_mascortos FROM samples.nyctaxi.trips WHERE samples.nyctaxi.trips.trip_distance > 0;
```

¿Cuál es el promedio de tarifa(fare_amount) y distanicia (trip_distance) de los viajes?

```python
%sql 
SELECT AVG(samples.nyctaxi.trips.fare_amount) AS Promedio_tarifa, AVG(samples.nyctaxi.trips.trip_distance) AS Promedio_distancia FROM samples.nyctaxi.trips;
```

Muestra los 20 viajes más recientes con su fecha , distancia y tarifa.

```python
%sql 
SELECT samples.nyctaxi.trips.tpep_pickup_datetime AS Fecha, 
samples.nyctaxi.trips.trip_distance AS Distancia, 
samples.nyctaxi.trips.fare_amount AS Tarifa 
FROM samples.nyctaxi.trips
ORDER BY samples.nyctaxi.trips.tpep_pickup_datetime DESC
LIMIT 20;
```

## Parte 2. Filtros y Condiciones

## Parte 3. Agregaciones
¿Cuántos viajes hay en total?

```python
%sql
SELECT COUNT(*)
FROM samples.nyctaxi.trips;
```

¿Cuál es el total de tarifas y la distancia total recorrida por todos los viajes?

```python
%sql
SELECT SUM(c.fare_amount) AS total_tarifas, SUM(c.trip_distance) AS total_distancia
FROM samples.nyctaxi.trips AS  c;
```

¿Cuál es la distancia promedio y la tarifa promedio de los viajes? (excluye valores cero)

```python
%sql 
SELECT ROUND(AVG(c.trip_distance),2) AS Distancia_promedio,
ROUNd(AVG(c.fare_amount),2) AS Tarifa_promedio
FROM samples.nyctaxi.trips c  
WHERE c.trip_distance > 0 
AND c.fare_amount > 0;
```

## Parte 4. GROUP BY
¿Cuántos viajes hay por zona de inicio (pickup_zip)? ¿Cuál es el ingreso total y promedio por zona? Ordena por ingreso total descendente.

```python
%sql
SELECT pickup_zip, 
COUNT(pickup_zip),
SUM(fare_amount) AS ingreso_total,
AVG(fare_amount) AS Ingreso_Promedio
FROM samples.nyctaxi.trips 
GROUP BY pickup_zip
ORDER BY ingreso_total DESC;
```

```python
%sql
SELECT
  pickup_zip,
  COUNT(*) as total_viajes,
  ROUND(SUM(fare_amount), 2) as ingresos_totales,
  ROUND(AVG(fare_amount), 2) as tarifa_promedio
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
ORDER BY ingresos_totales DESC;
```

¿Cuántos viajes hay por día de la semana? ¿Cuál es la distancia promedio por día?

```python
%sql
SELECT date_format(tpep_pickup_datetime,'EEEE') AS Dia_semana,
dayofweek(tpep_pickup_datetime) AS no_day,
COUNT(*) AS Total_viajes,
ROUND(AVG(trip_distance),2) AS Distancia_Promedio
FROM samples.nyctaxi.trips
WHERE trip_distance > 0
GROUP BY no_day, date_format(tpep_pickup_datetime,'EEEE')
ORDER BY no_day;
```

Cuántos viajes hay por hora del día? ¿Cuál es la tarifa promedio por hora? Ordena por hora.

```python
%sql
SELECT HOUR(tpep_pickup_datetime) AS Hora,
COUNT(*) AS Total_viajes,
ROUND(AVG(fare_amount),2) AS Tarifa_Promedio
FROM samples.nyctaxi.trips
GROUP BY Hora 
ORDER BY Hora ASC;
```

## Parte 5. HAVING
Muestra solo las zonas que tienen más de 1,000 viajes.

```python
%sql
SELECT pickup_zip AS Zona,
COUNT(*) AS Total_viajes
FROM samples.nyctaxi.trips
GROUP BY zona
HAVING Total_viajes >1000;
```

¿Qué días tuvieron más de 200 viajes? Muestra la fecha, cantidad de viajes y tarifa promedio.

```python
%sql
SELECT to_date(tpep_pickup_datetime) AS Fecha,
COUNT(*) AS Total_viajes, 
ROUND(AVG(trip_distance),2) AS Distancia_Promedio 
FROM samples.nyctaxi.trips
GROUP BY Fecha 
HAVING  Total_viajes > 200
ORDER BY Fecha ASC
LIMIT 20;
```

¿Qué horas tienen más de 1,000 viajes? Ordena por cantidad de viajes descendente.

```python
%sql
SELECT hour(tpep_pickup_datetime) AS hora,
COUNT(*) AS Total_viajes
FROM samples.nyctaxi.trips
WHERE trip_distance > 0
GROUP BY hora 
HAVING Total_viajes > 1000
ORDER by hora  DESC;
```
